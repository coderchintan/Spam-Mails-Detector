```python
base_path = '/home/chintan/Dataset/20newsbydate/20news-bydate-train/'
new_path = '/home/chintan/Dataset/20newsbydate/without_header/'
stop_path = '/home/chintan/Dataset/20newsbydate/stopwords_en.txt/'
base_path1 = '/home/chintan/Dataset/20newsbydate/20news-bydate-test/'
new_path1 = '/home/chintan/Dataset/20newsbydate/without_header_test/'



def hasNumbers(inputString):
    return bool(re.search(r'\d', inputString))
```

```python
#Reading Stop Words from a file
def find_stop_words(path):
    f = open(path,"r")
    words = f.read()
    f.close()
    return words.split()
```

```python
#Removing the file namded .DS_Store from each folder
def remove_file(base_path):
    folders = os.listdir(base_path)
    for folder in folders:
        file_path = base_path + folder
        files = os.listdir(file_path)
        for file in files:
            if file == '.DS_Store':
                os.remove(file_path + "/" + file)
    return "success"
```

```python
#Removing the Headers and Creating a new file with header removed in the new file
#approach used here is find the first line in the file which has "Lines:" String as the first word
#Extract number followed the "Lines:"
#Using these number we can remove the header by reading last n number of lines in the file
def remove_headers(base_path,new_path):
    folders = os.listdir(base_path)
    for folder in folders:
        file_path=  base_path  + folder
        files  = os.listdir(file_path)
        if not os.path.exists(new_path + folder):
            os.makedirs(new_path+folder)
        for file in files:
            num_of_lines = ['1']
            f = codecs.open(file_path + "/" + file,"r",encoding='utf-8',errors='ignore')
            lines = f.readlines()
            f.close()
            find_Lines_word = lines[:150]
            for line_word in find_Lines_word:
                words = line_word.split(":")
                if words[0]=='Lines':
                    num_of_lines = re.findall(r'\d+',line_word)
                    break
            if len(num_of_lines)>0:
                num_of_lines = (-int(num_of_lines[0]))
            else:
                num_of_lines = -1
                print(folder,file)
            new_lines = lines[num_of_lines:]
            fw = codecs.open(new_path + folder +  "/" + file,"w",encoding='utf-8',errors='ignore')
            for new_line in new_lines:
                fw.write(new_line)
            fw.close()
    return "MST"
```

```python
#Vocublary is created in such a way that stop words and the words which are not in english dictionary are removed
def create_vocabulary(new_path,stop_words,punct,diction):
    vocab  = []
    frequency = {}
    folders = os.listdir(new_path)
    for folder in folders:
        file_path = new_path + folder
        files = os.listdir(file_path)
        for file in files:
            f = codecs.open(file_path +  "/"+ file,"r",encoding='utf-8',errors='ignore')
            lines = f.readlines()
            f.close()
            for line in lines:
                words = line.split()
                for word in words:
                    word = word.lower()
                    word = word.strip()
                    word = word.strip(punct)
                    word = word.strip('!@#%&*()[]{}/?<>')
                    if hasNumbers(word):
                        continue
                    if len(word)<1:
                        continue
                    if word in stop_words:
                        continue
                    if len(word)<2:
                        continue
                    if '@' in word:
                        continue
                    if word.isdigit():
                        continue
                    if not diction.check(word):
                        continue
                    if word not in vocab:
                        vocab.append(word)
                    if word not in frequency:
                        frequency[word] = 0
                    frequency[word]+=1
    return vocab,frequency
```

```python
#Dataset is made to find the frequency of vocublary in each folder(target)
def create_dataset_required(new_path,final_sorted_list,stop_words,diction,punct):
    folders = os.listdir(new_path)
    dict1= {}
    for folder in folders:
        dict1[folder]={}
        file_path = new_path + folder
        files = os.listdir(file_path)
        print(folder)
        for file in files:
            f = codecs.open(file_path +  "/"+ file,"r",encoding='utf-8',errors='ignore')
            lines = f.readlines()
            f.close()
            for line in lines:
                words = line.split()
                for word in words:
                    word = word.lower()
                    word = word.strip()
                    word = word.strip(punct)
                    word = word.strip('!@#%&*()[]{}/?<>')
                    if hasNumbers(word):
                        continue
                    if len(word)<1:
                        continue
                    if word in stop_words:
                        continue
                    if len(word)<2:
                        continue
                    if '@' in word:
                        continue
                    if word.isdigit():
                        continue
                    if not diction.check(word):
                        continue
                    for vocab in final_sorted_list:
                        if vocab not in dict1[folder]:
                            dict1[folder][vocab] = 0
                        if word==vocab:
                            dict1[folder][vocab]+=1
                            break
    return dict1
```

```python
def predict(file_path,data_diction,target_files):
    dict1 = {}
    f = codecs.open(file_path,"r",encoding='utf-8',errors='ignore')
    list_words = f.read().split()
    for target in target_files:
        probab_word_log=0
        for word in list_words:
            word = word.lower()
            word = word.strip()
            word = word.strip(string.punctuation)
            word = word.strip('!@#%&*()[]{}/?<>')
            if word in data_diction[target]:
                probab_first = predict_function(data_diction,target,word)
                probab_word_log+=math.log2(probab_first)
        final_probability = probab_word_log * ((len(os.listdir('./20_newsgroups_train/' + target)))/16000)
        dict1[target] = final_probability
    return max(dict1.items(),key=operator.itemgetter(1))[0]
```

```python
def predict_function(data_diction,target,word):
    numerator = (data_diction[target][word])+1
    denominator = 0
    for i in data_diction[target]:
        denominator = denominator + data_diction[target][i]
    probab = numerator/(denominator + (1*(len(data_diction[target].keys()))))
    return probab
```

```python
def create_frame_for_test_for_accuracy(test_path):
    folders = os.listdir(test_path)
    frame_list = []
    for folder in folders:
        file_path = test_path + folder
        files = os.listdir(file_path)
        for file in files:
            small_list = []
            small_list.append(file)
            small_list.append(folder)
            frame_list.append(small_list)
    df = pd.DataFrame(frame_list,columns=['FileName','FileTarget'])
    return df
```
